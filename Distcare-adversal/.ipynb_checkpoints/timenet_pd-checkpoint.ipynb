```python
def get_loss(y_pred, y_true):
    loss = torch.nn.BCELoss()
    return loss(y_pred, y_true)
```

```python
def get_re_loss(y_pred, y_true):
    loss = torch.nn.MSELoss()
    return loss(y_pred, y_true)
```

```python
def get_kl_loss(x_pred, x_target):
    loss = torch.nn.KLDivLoss(reduce=True, size_average=True)
    return loss(x_pred, x_target)
```

```python
def get_wass_dist(x_pred, x_target):
    m1 = torch.mean(x_pred, dim=0)
    m2 = torch.mean(x_target, dim=0)
    v1 = torch.var(x_pred, dim=0)
    v2 = torch.var(x_target, dim=0)
    p1 = torch.sum(torch.pow((m1 - m2), 2))
    p2 = torch.sum(torch.pow(torch.pow(v1, 1/2) - torch.pow(v2, 1/2), 2))
    return torch.pow(p1+p2, 1/2)
```

```python
def pad_sents(sents, pad_token):

    sents_padded = []

    max_length = max([len(_) for _ in sents])
    for i in sents:
        padded = list(i) + [pad_token]*(max_length-len(i))
        sents_padded.append(np.array(padded))


    return np.array(sents_padded)
```

```python
def batch_iter(x, y, mask, lens, batch_size, shuffle=False):
    """ Yield batches of source and target sentences reverse sorted by length (largest to smallest).
    @param data (list of (src_sent, tgt_sent)): list of tuples containing source and target sentence
    @param batch_size (int): batch size
    @param shuffle (boolean): whether to randomly shuffle the dataset
    """
    batch_num = math.ceil(len(x) / batch_size) # 向下取整
    index_array = list(range(len(x)))

    if shuffle:
        np.random.shuffle(index_array)

    for i in range(batch_num):
        indices = index_array[i * batch_size: (i + 1) * batch_size] #  fetch out all the induces
        
        examples = []
        for idx in indices:
            examples.append((x[idx], y[idx], mask[idx], lens[idx]))
       
        examples = sorted(examples, key=lambda e: len(e[0]), reverse=True)
    
        batch_x = [e[0] for e in examples]
        batch_y = [e[1] for e in examples]
        batch_mask_x = [e[2] for e in examples]
#         batch_name = [e[2] for e in examples]
        batch_lens = [e[3] for e in examples]

        yield batch_x, batch_y, batch_mask_x, batch_lens
```

```python
def length_to_mask(length, max_len=None, dtype=None):
    """length: B.
    return B x max_len.
    If max_len is None, then max of length will be used.
    """
    assert len(length.shape) == 1, 'Length shape should be 1 dimensional.'
    max_len = max_len or length.max().item()
    mask = torch.arange(max_len, device=length.device,
                        dtype=length.dtype).expand(len(length), max_len) < length.unsqueeze(1)
    if dtype is not None:
        mask = torch.as_tensor(mask, dtype=dtype, device=length.device)
    return mask
```

```python
class TimeNet(nn.Module):
    def __init__(self, size, num_layers, dropout=0.0):
        super(TimeNet, self).__init__()
        self.size = size
        self.num_layers = num_layers
        self.dropout = dropout
        self.encoder = nn.GRU(input_size = 1, hidden_size = self.size, num_layers = self.num_layers, batch_first = True, dropout = self.dropout)
        self.decoder = nn.GRU(input_size = self.size, hidden_size = self.size, num_layers = self.num_layers, batch_first = True, dropout = self.dropout)
        self.Linear = nn.Linear(self.size, 1)


    def forward(self, x, lens):
        # Encoder
        # print(f'input.shape is {x.shape}')
        lens = lens.to('cpu')
        Batch, Timestep, features = x.shape
        outputs = torch.zeros((Batch, Timestep, features)).to(device)
        for i in range(features):
            encode , hidden_state = self.encoder(pack_padded_sequence(x[:,:,i].unsqueeze(-1), lens, batch_first=True))
    
            encode = pad_packed_sequence(encode,batch_first=True)[0]
            # print(f'hidden_states.shape is {hidden_state.shape}, encode.shape is {encode.shape}')
            # for j in range(Batch):
            #     start = 0
            #     end = lens[j] - 1
            #     while start < end:
            #         temp = encode[j, start, :]
            #         encode[j, start, :] = encode[j, end, :]
            #         encode[j, end, :] = temp
            #         start = start + 1
            #         end = end - 1
            # Decoder
            # output = torch.flip(hidden_states[-1], [0])
            # encode = torch.flip(encode, [1])

            decode, _ = self.decoder(pack_padded_sequence(encode, lens, batch_first=True))
            
            decode = pad_packed_sequence(decode, batch_first=True)[0]
            # print(f'decode.shape is {decode.shape}')
            output = self.Linear(decode)
            # print(f'output.device is {output.device}')
            outputs[:, :, i] = output.squeeze()
        # print(f'outputs.shape is {outputs.shape}')
        return outputs
```

```python
def get_n2n_data(x, y, x_len):
    length = len(x)
    assert length == len(y)
    assert length == len(x_len)
    new_x = []
    new_y = []
    new_x_len = []
    for i in range(length):
        for j in range(len(x[i])):
            new_x.append(x[i][:j+1])
            new_y.append(y[i][j])
            new_x_len.append(j+1)
    return new_x, new_y, new_x_len
```

```python
def transfer_gru_dict(pretrain_dict, model_dict):
    state_dict = {}
    for k, v in model_dict.items():
        if "encoder" in k:
            state_dict[k] = model_dict[k]
    return state_dict
```

```python
class TimeNet_Pretrain(nn.Module):
    def __init__(self, input_dim, hidden_dim, num_layers, dropout=0.0):
        super(TimeNet_Pretrain, self).__init__()
        self.input_dim = input_dim
        self.hidden_dim = hidden_dim
        self.num_layers = num_layers
        self.input_dim = input_dim
        self.dropout = dropout
        self.encoder = nn.GRU(input_size = 1, hidden_size = self.hidden_dim, num_layers = self.num_layers, batch_first = True, dropout = self.dropout)
        self.decoder = nn.GRU(input_size = self.hidden_dim, hidden_size = self.hidden_dim, num_layers = self.num_layers, batch_first = True, dropout = self.dropout)

        self.Linear1 = nn.Linear(self.num_layers * self.input_dim, 1)
        self.Linear2 = nn.Linear(self.hidden_dim, 1)
        self.Dropout = nn.Dropout(self.dropout)


    def forward(self, x, lens):
        # Encoder
        # print(f'input.shape is {x.shape}')
        lens = lens.to('cpu')
        Batch, _, features = x.shape
        outputs = torch.zeros((Batch, self.hidden_dim, self.input_dim * self.num_layers)).to(device)
        for i in range(features):
            encode , hidden_state = self.encoder(pack_padded_sequence(x[:,:,i].unsqueeze(-1), lens, batch_first=True))
            output = torch.permute(hidden_state,(1,2,0))
            outputs[:, :, self.num_layers*i: self.num_layers*(i+1)] = output.squeeze()
        # print(f'outputs.shape is {outputs.shape}')
        out_temp = self.Dropout(self.Linear1(outputs).squeeze())
        out = self.Linear2(out_temp)
        # print(f'out.shape is {out.shape}')
        return out
```

```python
def ckd_batch_iter(x, y, lens, batch_size, shuffle=False):
    """ Yield batches of source and target sentences reverse sorted by length (largest to smallest).
    @param data (list of (src_sent, tgt_sent)): list of tuples containing source and target sentence
    @param batch_size (int): batch size
    @param shuffle (boolean): whether to randomly shuffle the dataset
    """
    batch_num = math.ceil(len(x) / batch_size) # 向下取整
    index_array = list(range(len(x)))

    if shuffle:
        np.random.shuffle(index_array)

    for i in range(batch_num):
        indices = index_array[i * batch_size: (i + 1) * batch_size] #  fetch out all the induces
        
        examples = []
        for idx in indices:
            examples.append((x[idx], y[idx],  lens[idx]))
       
        examples = sorted(examples, key=lambda e: len(e[0]), reverse=True)
    
        batch_x = [e[0] for e in examples]
        batch_y = [e[1] for e in examples]
#         batch_name = [e[2] for e in examples]
        batch_lens = [e[2] for e in examples]
       

        yield batch_x, batch_y, batch_lens
```
